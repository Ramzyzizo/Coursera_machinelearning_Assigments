### ComputeCost

```python
def computeCost(x,y,theta):
    m=len(y)
    s=sum((np.dot(x,theta)-y))
    J=np.power(s,2)*(1/2/m)
    return J[0]
```

```python
x=np.array([[1,1],[1,2],[1,3]])
y=np.array([[1],[2],[3]])
theta=np.array([[0],[1]])
computeCost(x,y,theta)
```

### GradientDescent

```python
def gradientDescent(x,y,theta,alpha,i):
    m=len(y)
    J=np.zeros((i,1))
    for ii in range(1,i):
        error=np.dot(x,theta)-y
        theta=theta-np.dot(((alpha/m)*x).transpose(),error)
        J[ii]=computeCost(x,y,theta)
    return theta, J
```

```python
gradientDescent(x,y,theta,0.1,10)
```

### ComputeCostMulti

```python
def computeCostMulti(x,y,theta):
    m=len(y)
    s=sum(np.power((np.dot(x,theta)-y),2))
    J=np.power(s,2)*(1/2/m)
    return J[0]
```

```python
x=np.array([[1,1],[1,2],[1,3]])
y=np.array([[1],[2],[3]])
theta=np.array([[0],[1]])
computeCostMulti(x,y,theta)
```

### GradientDescentMulti

```python
def gradientDescentMulti(x,y,theta,alpha,i):
    m=len(y)
    J=np.zeros((i,1))
    for ii in range(1,i):
        error=np.dot(x,theta)-y
        theta=theta-np.dot(((alpha/m)*x).transpose(),error)
        J[ii]=computeCostMulti(x,y,theta)
    return theta, J
```

```python
gradientDescentMulti(x,y,theta,0.1,10)
```

### FeatureNormalize

```python
g=np.array([[1,1],[2,2],[3,3]])
```

```python
def featureNormalize(x):
    x_norm=x
    mu=np.zeros((1,np.size(x,1)))
    sigma=np.zeros((1,np.size(x,1)))
    mu=np.mean(g,axis=0)
    sigma=np.std(x,ddof=1,axis=0)
    for i in range(0,np.size(x,1)):
        for j in range(0,np.size(x,0)):
            x_norm[j,i]=(x[j,i]-mu[i])/sigma[i]
    return x_norm,mu,sigma
```

```python
featureNormalize(g)
```

### NormalEqn

```python
def normalEqn(x,y):
    A=np.linalg.inv(np.dot(x.transpose(),x))
    B=np.dot(x.transpose(),y)
    theta=np.dot(A,B)
    return theta
```
